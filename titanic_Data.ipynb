```python
import pandas as pd

def input_age(cols):
    Age = cols[0]
    Pclass = cols[1]

    # Check if Age is NaN
    if pd.isnull(Age):
        if Pclass == 1:
            return 37  # Replace with a value for Pclass 1
        elif Pclass == 2:
            return 29  # Replace with a value for Pclass 2
        elif Pclass == 3:
            return 24  # Replace with a value for Pclass 3
        else:
            return 30  # Default value for other cases

    return Age  # If Age is not NaN, return the existing value

# Assuming you have a DataFrame 'df' with 'Age' and 'Pclass' columns
data['Age'] = data[['Age', 'Pclass']].apply(input_age, axis=1)
```

```python
sns.heatmap(data.isnull() , yticklabels=False , cbar=False , cmap='viridis')
```

```python
import pandas as pd

def impor_cabin(maker):
    Cabin = maker[0]
    Pclass = maker[1]

    if  pd.isnull(Cabin):
        if Pclass == 1 :
            return 20
        elif Pclass == 2:
            return 30
        
        elif Pclass == 3:
            return 24
    return Cabin

data['Cabin'] = data[['Cabin' , 'Pclass']].apply(impor_cabin , axis = 1)
```

```python
sns.heatmap(data.isnull() , yticklabels=False , cbar=False, cmap='viridis' )
```
